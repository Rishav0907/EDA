# __Q-Q Plots__

```python
import numpy as np
import scipy.stats as stats
import pylab
```

```python
#generating sample of random numbers 
std_normal=np.random.normal(loc=0,scale=1,size=1000)
# loc = mean 
# scale= standard devaition
# size = number of points
for i in range(0,101): # 0 to 100th percentile
    print(i,np.percentile(std_normal,i))
```

```python
measurements=np.random.normal(loc=20,scale=5,size=100)
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```

```python
# increasing the number of points to 1000
measurements=np.random.normal(loc=20,scale=5,size=1000)
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```

```python
measurements=np.random.uniform(low=-1,high=1,size=1000)
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```
