# __Box Plots and Whiskers__

```python
# Box plots are another method to plot a 1D plot
# It uses the concept of mean,median and quantiles
# It displays the 25th,50th and 75th percentile
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

iris=pd.read_csv('iris.csv')
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
plt.show()
```
