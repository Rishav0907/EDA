# __Pair-Plots__

```python
# Pair plots are unique scatter plots of each pair of data 
# here in this iris data we have 4 classifications and we want to plot data on two
# So possible number of plots is 4C2 ie 6
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd

iris=pd.read_csv('iris.csv')
```

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris , hue='species' , height=3)
plt.show()

# The diagonal plots are about pdf and cdn so not considering at the moment.
# The plots above the diagonal and below the diagonal are mirror images.
# So considering the above diagonal plots
# comparing the above  plots it can be seen that the last plot with petal lengths and width as x and y axis 
# the flower are more seperated than others. 
```

## __Considering last of the above  plots__

### we can generate the follwing algorithm
#### if petal_width <=1 and petal_length<=2 then it is a setosa flower
#### if petal_width>=1 and petal_width<2 and pair_length >=4 and pair_length>5 then it is versicolor
