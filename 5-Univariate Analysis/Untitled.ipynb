# __Univariate Analysis__

#### As the name suggests univariate analysis means analysis of a single variable or feature. Here we will be analysing all the features one by one like petal width and length ,sepal width and length

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd

iris=pd.read_csv('iris.csv')
```

```python
sns.FacetGrid(iris , hue='species' , height=5) \
   .map(sns.distplot,'petal_length') \
   .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris , hue='species' , height=5) \
   .map(sns.distplot,'petal_width') \
   .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris , hue='species' , height=5) \
   .map(sns.distplot,'sepal_length') \
   .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris , hue='species' , height=5) \
   .map(sns.distplot,'sepal_width') \
   .add_legend()
plt.show()
```

### Observations:
#### My order of preference will be petal_length > > petal_width > sepal_length > petal_width
