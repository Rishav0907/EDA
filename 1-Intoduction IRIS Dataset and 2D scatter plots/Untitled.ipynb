```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd

iris=pd.read_csv('iris.csv')
```

```python
# Q1) How many data points and features?
iris.shape
```

```python
# Q2) How many columns are available?
iris.columns
```

```python
# Q3) How many data points for each class are present?
iris['species'].value_counts() # To find how mnay plants of a particular species are present
# balanced dataset vs imbalanced dataset
# balanced dataset: One in which the data amounts in each class is amost same
# imbalanced dataset: data amount is not same
# This is a balanced data set
```

## __2D Scatter Plot__

```python
# 2-D scatter plot
```

```python
iris.plot(kind='scatter' , x='sepal_length' , y='sepal_width')
plt.show()
# cannot be distinguished as all of them are having the same colour
# lets color them out
```

```python
# 2-D scatter plot with color coding of each plant
sns.set_style('whitegrid')
sns.FacetGrid(iris,hue='species',size=5) \
         .map(plt.scatter , 'sepal_length' , 'sepal_width') \
         .add_legend()
plt.show()
# Here the blue points are easily seperable from the other points
# but the orange and green  points cannot be seperated from each other
```

### __Observations :__

#### 1. Using sepal_length and sepal_width features we can easily seperate the Setosa flowers from each other
#### 2. Seperating the Versicolor and Verginica is comparatively very difficult
