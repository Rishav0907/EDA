# __CUMILATIVE DENSITY FUNCTION (CDF)__

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

iris=pd.read_csv('iris.csv')
```

```python
iris_setosa      = iris.loc[iris['species']=='Setosa']
iris_virginica   = iris.loc[iris['species']=='Virginica']
iris_versicolor  = iris.loc[iris['species']=='Versicolor']
```

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```

### __Observation:__
#### 1. PDF measures how many petal lengths of a particular length are present. For eg: b/w 1.5 nd 1.6 it will measure how many flowers have petal length between 1.5 & 1.6
#### 2. CDF measures a complete different thing. I tmeasures how many flowers have petal_length less than or equal to a particular length.
#### 3. Thus the area under a particular point of pdf gives the cdf
#### 4. Differntiating cdf gives pdf 
#### 5. Integrating pdf gives cdf

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
sns.set_style('whitegrid')
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(iris_virginica['petal_length'], bins=10, density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
sns.set_style('whitegrid')
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins=10, density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
sns.set_style('whitegrid')
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

### __Observations:__
#### We can make the follwing conditions:
#### 1. if petal_length < 2 then setosa flower
#### 2. Considering petal_length 5 as threshold
#### 3. if petal_length >2 and <=5 then virginica (95% correct predictions)
#### 4. if petal_length >5 then versicolor (10% wrong predictions)
