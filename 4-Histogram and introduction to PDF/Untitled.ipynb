# __Histogrma , PDF , CDF__

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
iris=pd.read_csv('iris.csv')

# What about 1-D plots?
```

```python
iris_setosa      = iris.loc[iris['species']=='Setosa']
iris_virginica     = iris.loc[iris['species']=='Virginica']
iris_versicolor  = iris.loc[iris['species']=='Versicolor']
# 1D plot of petal length
sns.set_style('whitegrid')
plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa['petal_length']), 'o')
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor['petal_length']), 'o')
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica['petal_length']), 'o')
plt.xlabel('petal_length')
plt.ylabel('count')
plt.show()

# Very hard to differentiate all the points
# Does not makes sense
```

```python
# Makig a histogram makes more sense and details than a 1d scatter plot
sns.FacetGrid(iris,hue='species',height=5) \
    .map(sns.distplot, 'petal_length') \
    .add_legend();
plt.show()
# Blue is the histogram of petal lengths of setosa
# Orange is for versicolor
# green is for virginica
# y is not a different parameter , its jst a count of the petal lengths
# the smooth lines are Probability density functions(PDF)
```

### __From the above histogram we can form the follwing algorith:__

#### if petal_length <=2 then setosa
#### if petal_length>2 and petal_length<4 then versicolor
#### else virginica
